```python
import pandas as pd
from pathlib import Path

path = Path('../data/synthetic_session_panel.csv')
df = pd.read_csv(path, parse_dates=['quote_ts_utc','quote_ts_local','date'])
print(df.shape)
df.head()
```

```python
key_cols = ['effective_price','conversion','market_id','date']
missing = df[key_cols].isna().mean().rename('missing_rate')
missing
```

```python
recon = (df['item_subtotal'] + df['delivery_fee'] + df['service_fee'] + df['surge_fee'] - df['discount_total'])
max_abs_diff = (recon - df['effective_price']).abs().max()
print('max add-up diff:', float(max_abs_diff))
```
