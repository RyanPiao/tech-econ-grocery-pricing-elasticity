# Step 4 Baseline Econometric Model

Implements staggered DiD skeleton with first-stage and reduced-form estimates.

```python
import pandas as pd
import statsmodels.formula.api as smf

df = pd.read_csv('../data/synthetic_session_panel.csv', parse_dates=['quote_ts_local'])
df = df[df['contaminated'] == 0].copy()
print(df.shape)
```

```python
fs = smf.ols('ln_effective_price ~ did_treat_post + eta_minutes + stockout_rate + rain_index + C(market_id) + C(stage) + C(hour_local)', data=df).fit(cov_type='cluster', cov_kwds={'groups': df['market_id']})
rf = smf.ols('conversion ~ did_treat_post + eta_minutes + stockout_rate + rain_index + C(market_id) + C(stage) + C(hour_local)', data=df).fit(cov_type='cluster', cov_kwds={'groups': df['market_id']})
print('FS did_treat_post:', fs.params['did_treat_post'])
print('RF did_treat_post:', rf.params['did_treat_post'])
print('Implied elasticity:', rf.params['did_treat_post']/fs.params['did_treat_post'])
```

```python
fe = smf.ols('conversion ~ ln_effective_price + eta_minutes + stockout_rate + rain_index + C(market_id) + C(stage) + C(hour_local)', data=df).fit(cov_type='cluster', cov_kwds={'groups': df['market_id']})
fe.summary().tables[1]
```
