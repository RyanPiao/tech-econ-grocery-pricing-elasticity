# Step 3 EDA Notebook

Explores distributions, market-level summaries, and conversion-price patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/synthetic_session_panel.csv', parse_dates=['quote_ts_local'])
print(df.shape)
```

```python
market_summary = df.groupby('market_id').agg(
    sessions=('session_id','count'),
    conversion_rate=('conversion','mean'),
    mean_effective_price=('effective_price','mean'),
    contaminated_rate=('contaminated','mean')
).reset_index()
market_summary
```

```python
dec = df.copy()
dec['price_decile'] = pd.qcut(dec['effective_price'], 10, labels=False, duplicates='drop') + 1
plot_tab = dec.groupby('price_decile').agg(conversion_rate=('conversion','mean'), mean_effective_price=('effective_price','mean')).reset_index()
plt.plot(plot_tab['mean_effective_price'], plot_tab['conversion_rate'], marker='o')
plt.xlabel('Mean effective price')
plt.ylabel('Conversion rate')
plt.title('Conversion vs price deciles')
plt.grid(alpha=0.2)
```
